```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from hypernetworks.utils.HTTools import get_subHn_by_semantic_boundary

from IPython.display import Image, display

parser = load_parser()
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
        Car=<Engine, Gears, Wheels; R_car; B(Car)>
        Engine=<engine, ignition, key, revs; R_engine; B(Engine)>
            engine={engineOff, engineOn}
            key={keyIn, keyOut}
            ignition={offPosition, onPosition}

        Gears=<clutch, gearbox, gearstick; R_gears; B(Gears)>
            gearbox={parked, neutral, reverse}
            gearstick={parkSelected, neutralSelected, driveSelected, reverseSelected}
            clutch={clutchEngaged, clutchDisengaged}

        Wheels=<wheel, break; R_wheels; B(Wheels)>
            wheel={stopped, moving}
            break={breakApplied, applying, breakReleased}
""")

# print(hn)

draw_hn(hn, fname="pictures/car-hn", view=False)
Image(filename='pictures/car-hn.png', width=1600, height=400)
```
