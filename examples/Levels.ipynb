```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from hypernetworks.utils.HTSearch import what_is_it, bottom_up, get_search_paths, top_down
from hypernetworks.utils.HTTools import get_subHn_by_semantic_boundary, get_subHn_by_level

from IPython.display import Image, display

parser = load_parser()
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    y=<a, b; R>^N(1)
    a=<x1, x2, x3, y2; R>^N(0)
    b=<y1, y2; R>^N(-1)
""")

draw_hn(hn, fname="/tmp/levels-hn", view=False, split_camel=False, show_levels=True, direction="TB")
Image(filename='/tmp/levels-hn.png', width=400, height=300)
```

```python
level_hn = Hypernetwork()

level_hn = get_subHn_by_level(hn, "N+1", level_hn)

draw_hn(level_hn, fname="/tmp/level-hn", view=False, split_camel=False, show_levels=False, show_vertex=True, direction="TB")
Image(filename='/tmp/level-hn.png', width=200, height=300)
```
