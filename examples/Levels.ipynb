```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from hypernetworks.utils.HTSearch import what_is_it, bottom_up, get_search_paths, top_down
from hypernetworks.utils.HTTools import get_subHn_by_semantic_boundary, get_subHn_by_level

from IPython.display import Image, display

parser = load_parser()

pictures_dir = "pictures/levels/"
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    y=<a, b; R>[N+1]
    a=<x1, x2, x3, y2, c; R>[N]
    b=<y1, y2, c, d; R>[N0]
    c=<z1, z2, z3, z4, d; R>[N]
    d=<w1, w2; R>[N-1]
""")

draw_hn(hn, fname=pictures_dir+"levels-hn", view=False, split_camel=False, show_levels=True, show_vertex=False, direction="TB")
Image(filename=pictures_dir+"levels-hn.png", width=400, height=300)
```

```python
level_hn = Hypernetwork()

level_hn = get_subHn_by_level(hn, "N+1", level_hn)

draw_hn(level_hn, fname=pictures_dir+"level-sub-hn1", view=False, split_camel=False, show_levels=False, show_vertex=False, direction="TB")
Image(filename=pictures_dir+"level-sub-hn1.png", width=100, height=300)
```

```python
level_hn = Hypernetwork()

level_hn = get_subHn_by_level(hn, "N", level_hn)

draw_hn(level_hn, fname=pictures_dir+"level-sub-hn2", view=False, split_camel=False, show_levels=False, show_vertex=False, direction="TB")
Image(filename=pictures_dir+"level-sub-hn2.png", width=400)
```
