```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA, hstype_to_str
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTTools import remove_special, is_seq
from hypernetworks.utils.HTGraph import draw_hn

from IPython.display import Image
```

```python
hn = Hypernetwork()
parser = load_parser()

compile_hn(hn, parser, """
    a=<x1, x2, x3; R>
    b=<y2, y2, y3, y4; R>
""")

draw_hn(hn, fname="/tmp/hn", view=False)
Image(filename='/tmp/hn.png', width=600, height=600)
```

```python
new_hn = Hypernetwork()

compile_hn(new_hn, parser, """
    c=<a, x2, x3, y1, y2; R>
""")

draw_hn(new_hn, fname="/tmp/hn", view=False)
Image(filename='/tmp/hn.png', width=600, height=600)
```

```python
hn.union(new_hn)

draw_hn(hn, fname="/tmp/hn", view=False)
Image(filename='/tmp/hn.png', width=800, height=1200)
```

Hyperunion *hn* with *new_hn* with hn.union(new_hn) 

```python
new_hn = Hypernetwork()

compile_hn(new_hn, parser, """
    b=<y1, y2, y3; R>
""")

hn.union(new_hn)

draw_hn(hn, fname="/tmp/hn", view=False)
Image(filename='/tmp/hn.png', width=800, height=1200)
```

Creates a new version of *b*, but because one already exists the original and the new are added under a $\beta$-aggregation
