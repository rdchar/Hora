## Simple Example

#### Initialise the parser and load the libraries

```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from IPython.display import Image

parser = load_parser()
```

#### Create a hypernetwork

```python
first_hn = Hypernetwork()

compile_hn(first_hn, parser, """
    a=<x1, x2, x3; R>
    b=<y1, y2, y3, y4; R>
""")

draw_hn(first_hn, fname="/tmp/hn1", view=False)
Image(filename='/tmp/hn1.png', width=500)
```

#### Create and merge a second hypernetwork

```python
second_hn = Hypernetwork()

compile_hn(second_hn, parser, """
    c=<a, x2, x3, y1, y2; R>
""")

first_hn.merge(second_hn)

draw_hn(first_hn, fname="/tmp/hn2", view=False)
Image(filename='/tmp/hn2.png', width=500)
```

#### Merge the two hypernetworks

```python
first_hn.merge(second_hn)

draw_hn(first_hn, fname="/tmp/hn3", view=False)
Image(filename='/tmp/hn3.png', width=600)
```
