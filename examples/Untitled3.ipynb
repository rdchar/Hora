```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import ALPHA
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn

from IPython.display import Image
```

```python
parser = load_parser()

hn = Hypernetwork()

compile_hn(hn, parser, """
    a=<x1, x2, b>
    b=<a, x2, x3>
""")

hn.insert(vertex="','.expression+", hstype=ALPHA, simplex=["','.expression+", "'z'"])

print(hn)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    alpha=<x_1, x_2, x_3; R>
""")

draw_hn(hn, fname="/tmp/alpha", view=True)
Image(filename='/tmp/alpha.png', width=300, height=1200)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    beta={x_1, x_2, x_3; R}
""")

draw_hn(hn, fname="/tmp/beta", view=True)
Image(filename='/tmp/beta.png', width=300, height=1200)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    w=<a, x=<b, c, e>, y=<d, z>, z>
    z=<e, f, g>
""")

draw_hn(hn, fname="/tmp/hs-1", view=True)
Image(filename='/tmp/hs-1.png', width=300, height=1200)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    w=<a, x, y, z>
    x=<b, c, e>
    y=<d, z>
    z=<e, f, g>
""")

draw_hn(hn, fname="/tmp/hs-2", view=True)
Image(filename='/tmp/hs-2.png', width=300, height=1200)
```
