```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from hypernetworks.core.HTErrors import HnSearchError, HnVertexNoFound
from hypernetworks.utils.HTMatrix import to_matrix, from_matrix, matrix_to_string

import logging as log
import numpy as np
from pprint import pprint

from IPython.display import Image

parser = load_parser()
```

**Note**: This essentially this creates a *symetric* incident matrix from which it generates this *Hn*.

```python
friends_1 = Hypernetwork()

compile_hn(friends_1, parser, """
    a=<abby,     bess,     cody, not_dana; R_friends>
    c=<abby,     bess,     cody, dana; R_friends>
    b=<not_abby, bess,     cody, not_dana; R_friends>
    d=<not_abby, not_bess, cody, dana; R_friends>
""")

draw_hn(friends_1, fname="/tmp/friends-1", view=False)
Image(filename='/tmp/friends-1.png', width=700, height=500)
```

```python
M = to_matrix(friends_1, R="friends")
# pprint(M)
new_friends_1 = from_matrix(M, R="friends")

draw_hn(new_friends_1, fname="/tmp/friends-M-1", view=False)
Image(filename='/tmp/friends-M-1.png', width=1000, height=600)
```

```python
M4 = [['a', 'b', 'c', 'd', 'e', 'f', 'g'],
      ['a', 'h', 'c', 'd', 'e', 'i', 'g']]

hn_M4 = from_matrix(M4, replace=False)
draw_hn(hn_M4, fname="/tmp/hn-M4", view=False)
Image(filename='/tmp/hn-M4.png', width=500, height=500)
```
