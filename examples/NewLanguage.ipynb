```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn

from hypernetworks.utils.HTSpaces import get_space
from hypernetworks.utils.HTSearch import what_is_it

from IPython.display import Image, display

parser = load_parser()
```

The programming language:



```python
chn_language = Hypernetwork()

compile_hn(chn_language, parser, """
    Program=<~NAME, backcloths=(Backcloth), modules=(Module), boundaries=(Boundary)>
    Module=<~NAME, functions=(Functions)>
    Backcloth=<~NAME, hypersimplicies=(Hypersimplex)>
    Hypersimplex={Alpha, UnionAlpha, Beta, Vertex, Sequence}
    Alpha=<~NAME, simplex=(X_Hypersimplex), Relation, time, PsiFunction, PhiFunction, Boundary>
    UnionAlpha=<~NAME, simplex=(X_Hypersimplex), Relation, time, PsiFunction, PhiFunction, Boundary>
    Beta=<~NAME, simplex=(X_Hypersimplex), Relation, time, PsiFunction, PhiFunction, Boundary>
        Vertex=<~NAME>
        Relation=<~NAME, comprehension>
        Boundary=<~NAME, X>
        PsiFunction=<~NAME>
        PhiFunction=<~NAME>
        Sequence=<~NAME, X_Hypersimplex>
    Functions={Function, PsiFunction, PhiFunction}
        Function=<~NAME, params=(Param), block=(Block)>
        Param=<~NAME, Type>
        Selection=<Condition, X_Block>
            Block={X_Function, Selection}
    Type={INT, STR, Hypersimplex}
    Library={Search}
""")

draw_hn(chn_language, fname="/tmp/CHn-new-language", view=False, split_camel=False)
Image(filename='/tmp/CHn-new-language.png', width=1200, height=1200)
```
