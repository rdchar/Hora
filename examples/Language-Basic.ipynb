```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA, hstype_to_str
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTTools import remove_special
from hypernetworks.utils.HTGraph import draw_hn

from IPython.display import Image

parser = load_parser()

pictures_dir="pictures/language/"
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    ς=<x_1, x_2, x_3; R>
""")

draw_hn(hn, fname=pictures_dir+"alpha", view=False)
Image(filename=pictures_dir+"alpha.png", width=300, height=1200)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    ς={x_1, x_2, x_3; R}
""")

draw_hn(hn, fname=pictures_dir+"beta", view=False)
Image(filename=pictures_dir+"beta.png", width=300, height=1200)
```

```python
hn = Hypernetwork()

compile_hn(hn, parser, """
    w=<a, x=<b, c, e; R>, y=<d, z; R>, z; R>
    z=<e, f, g; R>
""")

draw_hn(hn, fname=pictures_dir+"hs-1", view=False)
Image(filename=pictures_dir+"hs-1.png", width=400, height=1200)
```

Given this and the next Hs we look to find the meet of the two Hn's.  The approach with Hora has been to create a new Hn that contains all the common parts.  

This is the first of the two Hn's, $Hn_1$.

```python
hn1 = Hypernetwork()

compile_hn(hn1, parser, """
    w=<a, x, y, z>
    x=<b, c, e>
    y=<d, z>
    z=<e, f, g>
""")

draw_hn(hn1, fname=pictures_dir+"hs-1", view=False)
Image(filename=pictures_dir+"hs-1.png", width=400, height=1200)
```
