```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn
from hypernetworks.utils.HTSpaces import get_space
from hypernetworks.utils.HTSearch import what_is_it

from IPython.display import Image, display

pictures_dir = "pictures/banking/"
parser = load_parser()
```

```python
banking_hn = Hypernetwork()

compile_hn(banking_hn, parser, """
    has=<customer, %1, account, %n; R_association>
    associated-with=<account, %1, transaction, %n; R_association>
""")

draw_hn(banking_hn, fname=pictures_dir+"banking-hn", show_vertex=False, view=False)
Image(filename=pictures_dir+'banking-hn.png', width=500, height=1200)
```

```python
banking_hn = Hypernetwork()

compile_hn(banking_hn, parser, """
    customer=<%ID, name, address, has=(account); R_customer>
    account=<%ID, opened, associated-with=(transaction); R_account>
    transaction=<payee, %date, %amount; R_transaction>
""")

draw_hn(banking_hn, fname=pictures_dir+"banking-2-hn", show_vertex=False, view=False)
Image(filename=pictures_dir+'banking-2-hn.png', width=500, height=1200)
```
