```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn

from IPython.display import Image, display

parser = load_parser()
```

```python
topics_hn = Hypernetwork()

compile_hn(topics_hn, parser, """
    CHn=<CHnHypernetwork, Concepts, Search, Algebra>
    Hypernetwork=<Backcloth, Traffic>
    CHnHypernetwork=<Hypernetwork, CHnBackcloth>
    Backcloth=<Hypersimplex={alpha, beta}, Vertex={vertex, property}, Rels, Time>
    CHnBackcloth=<Backcloth, Boundaries, Levels, Sequences>
    Rels={implicit, explicit}
    Traffic=<Coloured, traffic, Functions=<psi, psi^-1, phi, phi^-1>>
    Concepts={Overlapping, Peaks, Paths, R-analysis, Flexibility, Mutability, Logic}
    Algebra={HUnion, HIntersection, HMember, Sub-Hn, EmptyHn}
    Search={}
    AbstractConcept=<Semantics, Models={Abstract=<Backcloth>, Concrete=<Backcloth, Traffic>, Dynamics=<Traffic>}>
    UseCases={}
""")

draw_hn(topics_hn, fname="/tmp/Thesis-topics-hn", view=False, split_camel=False)
Image(filename='/tmp/Thesis-topics-hn.png', width=2000, height=1200)
```

```python
implementation_hn = Hypernetwork()

compile_hn(implementation_hn, parser, """
    Hora=<DataStructure, CoreLibraries, SupportingLibraries, Compiler>
    SupportingLibraries=<Search, Paths, Graph, Matrix, SimplicalComplex, Spaces>
    CoreLibraries=<Hypernetwork, Hypersimplex>
""")

draw_hn(implementation_hn, fname="/tmp/Thesis-implementation-hn", view=False)
Image(filename='/tmp/Thesis-implementation-hn.png', width=2000, height=1200)
```

```python
chapters_hn = Hypernetwork()

compile_hn(chapters_hn, parser, """
    Thesis=<Introduction, LitReview, Problem, Foundations, CompHn, Examples, Discusion, FutherWork>
    LitReview=<ModellingConcepts, Comparison={OO, DataLanguages={XML-JSON}, SemanticWeb, DDD, KnowledgeGraph, Networks, LinkedData}>
    Foundations=<Hypernetwork>
    CompHn=<CHn>
    Examples=<SharingCities, MEDA, HyperPn>
        SharingCities=<Traffic>
        MEDA=<Hora, HnBackcloth, Search>
        HyperPn=<Hora>
""")

draw_hn(chapters_hn, fname="/tmp/Thesis-chapters-hn", view=False)
Image(filename='/tmp/Thesis-chapters-hn.png', width=2000, height=1200)
```

```python
thesis_hn = Hypernetwork()

thesis_hn.union(topics_hn)
thesis_hn.union(implementation_hn)
thesis_hn.union(chapters_hn)

draw_hn(thesis_hn, fname="/tmp/Thesis-thesis-hn", view=False)
Image(filename='/tmp/Thesis-thesis-hn.png', width=2000, height=1200)
```
