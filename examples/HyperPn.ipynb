```python
from hypernetworks.core.Hypernetwork import Hypernetwork
from hypernetworks.core.Hypersimplex import VERTEX, ALPHA, BETA, hstype_to_str
from hypernetworks.utils.HTCompiler import load_parser, compile_hn
from hypernetworks.utils.HTGraph import draw_hn

from IPython.display import display, Javascript, HTML, Image

parser = load_parser()
```

```python
hpn_1 = Hypernetwork()

compile_hn(hpn_1, parser, """
    hpn_1={p1_1, p1_2, p1_3, mutex_1; B(hpn_1)}
""")

draw_hn(hpn_1, fname="/tmp/hpn_1", view=False, direction="LR")
Image(filename='/tmp/hpn_1.png', width=200, height=1200)
```

```python
hpn_2 = Hypernetwork()

compile_hn(hpn_2, parser, """
    hpn_2={p2_1, p2_2, p2_3, mutex_1; B(hpn_2)} 
""")

draw_hn(hpn_2, fname="/tmp/hpn_2", view=False, direction="LR")
Image(filename='/tmp/hpn_2.png', width=200, height=1200)
```

```python
hpn = Hypernetwork()

compile_hn(hpn, parser, """
    hpn=<hpn_1, hpn_2>
    hpn_1={p1_1, p1_2, p1_3, mutex_1; B(hpn_1)}
    hpn_2={p2_1, p2_2, p2_3, mutex_1; B(hpn_2)} 
""")

res = set()
for key, hs in hpn.hypernetwork.items():
    if len(hs.B) > 1:
        res.add(key)
        
        print(key, type(hs.B))
        
print(res)
```
